# Importing the neccesary libraries

```python
import pdfplumber as pdf
import pandas as pd
import nltk
import re
```

# opening a file and converting the pdf to text using pdfplumber


```python
x=["CV_NIAPOLICYSCHEDULECIRTIFICATECV_89461004.pdf","CV_NIAPOLICYSCHEDULECIRTIFICATECV_89461004.pdf","CV_NIAPOLICYSCHEDULECIRTIFICATECV_89673497.pdf","CV_NIAPOLICYSCHEDULECIRTIFICATECV_92613728.pdf","CV_NIAPOLICYSCHEDULECIRTIFICATECV_92650318.pdf"]
file=open(x[4],"rb")
```

```python
pdf_new=pdf.open(file)
text=""
for i in range(len(pdf_new.pages)):
    page=pdf_new.pages[i]
    text+=page.extract_text()
print(text)


    



```

 # Parsing using regex to obtain the neccesary outputs and a sample output is displayed


```python
find_policy=re.compile(r"\d{20}")
policy_number=pd.Series(find_policy.search(text).group())
policy_number
find_name=re.compile(r"Name [A-Z,\.,\s,\(cid:9\)]*\s")
name=pd.Series(find_name.findall(text)[0].split(maxsplit=1)[1])
print(name)
find_number=re.compile(r"Contact Number  \/  \/[\d,\s]*" )
number=pd.Series(find_number.findall(text)[0].split()[-1])
print(number)
find_contact=re.sub(r"Contact Number  \/  \/[\d,\s,\n]*"," ",text)
find_Address=re.compile(r"Address [A-Z,\.,\d,/,\s,\n,\(cid:9\)]*",re.MULTILINE)
address=pd.Series(find_Address.findall(find_contact)[0].split(maxsplit=1)[-1])
print(address)

find_type=re.compile(r"Commercial .*?(?=S)")
types=pd.Series(find_type.findall(text)[0].split(maxsplit=1)[-1])
types
find_regno=re.compile(r"no\. .*")
register_num=pd.Series(find_regno.findall(text)[0].split(".",maxsplit=1)[1])
register_num
find_amount=re.compile(r"Total Payable .+")
amount=pd.Series(find_amount.findall(text)[0].split()[-1])
amount
find_customerid=re.compile(r"ID [A-Z,a-z,\(,\),:,0-9,\s]*\)")
customer_id=pd.Series(find_customerid.findall(text))
print(customer_id)
find_gvw=re.compile(r"Gross Vehicle Weight \d*")
gvw=pd.Series(find_gvw.findall(text)[0].split()[-1])
print(gvw)
data=pd.concat([policy_number,customer_id,name,address,number,types,gvw,register_num,amount],axis=1)
data.head()






```

# Writing the data into csv file

```python
data.to_csv(r'data_mining.csv',mode="a",header=None,index=False)

```
